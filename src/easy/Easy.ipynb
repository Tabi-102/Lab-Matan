```python
#Easy2
import numpy as np
import matplotlib.pyplot as plt

def f(x, r):
    return r * x * (1 - x)

x = np.linspace(0, 1, 400)
rs = [0.5, 1.0, 2.5, 3.2, 3.8]

plt.figure(figsize=(6, 4))
for r in rs:
    plt.plot(x, f(x, r), label=f"r = {r}")

plt.plot(x, x, linestyle="--", label="y = x")  # y = x
plt.xlabel("x_n")
plt.ylabel("x_{n+1}")
plt.legend()
plt.grid(True)
plt.title(" x_{n+1} = r x_n (1 - x_n)   r")
plt.tight_layout()
plt.show()
```

```python
#Easy3
import numpy as np
import matplotlib.pyplot as plt

def g(x, r):
    return r * x * (1 - x) * (2 - x)

x = np.linspace(0, 1, 400)

#r = [0, 3*sqrt(3)/2]  [0, 2.598]
rs = [0.5, 1.0, 2.5, 3.2, 3.8]

plt.figure(figsize=(6, 4))
for r in rs:
    plt.plot(x, g(x, r), label=f"r = {r}")

plt.xlabel("x_n")
plt.ylabel("x_{n+1}")
plt.title(" x_{n+1} = r x_n(1-x_n)(2-x_n)   r")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
